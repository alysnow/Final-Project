```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
import os
```

```python
 # Read the csv file into a pandas DataFrame
data = pd.read_csv("../Resources/data/IMDB-Movie-Data.csv")
data.head()
```

```python
movie_df = data[['Genre1','Genre2','Genre3','Rating','Votes','Metascore','Runtime (Minutes)','Revenue (Millions)']]
```

```python
movie_df = movie_df.dropna()
movie_df
```

```python
movie_df['word_rating'] = np.where(movie_df['Rating'] >6, "Good", "Bad")
```

```python
movie_df.info()
```

```python
movie_df
```

```python
# X Values df
new_movie_df = movie_df[['Genre1','Genre2','Genre3','Metascore']]
```

```python
new_movie_df.dtypes
```

```python
data['Genre1'].value_counts()
```

```python
# Assign the data to X
X = pd.get_dummies(data=new_movie_df,columns=['Genre1','Genre2','Genre3'])
X
```

```python
# Assign the data to y (47 Variables which are going to predict the Revenue)
y = movie_df["word_rating"]
```

```python
# Use train_test_split to create training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Support vector machine linear classifier
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
# Model Accuracy
print('Test Acc: %.3f' % model.score(X_test, y_test))
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
predictions = model.predict(X_test)
print(classification_report(y_test, predictions))
```
